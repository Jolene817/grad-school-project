```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
df = pd.read_csv('data/Train.csv')
df.head()
```

```python
df.info()
```

```python
df['saledate'] = df['saledate'].astype('datetime64[ns]')
```

```python
df['age'] = df.saledate.dt.year - df.YearMade
```

```python
df['SalesYear'] = df.saledate.dt.year
```

```python
df = df.join(pd.get_dummies(df.ProductGroup))
```

```python
df['is_Vehicle'] = pd.get_dummies(df['MachineHoursCurrentMeter'] > 0)[1]
```

```python
plt.scatter(df.SalePrice, df.ProductGroup)
```

```python
df['ProductGroupRank'] =  df['ProductGroup'].map({'MG': 1, 'TTT': 2, 'TEX': 3, 'WL': 4, 'BL': 5, 'SSL': 6})
```

```python
df.Enclosure.value_counts()
```

```python
pd.get_dummies(df['Enclosure'])
enclosure_dummy = pd.get_dummies(df['Enclosure'])
df[ ['OROPS','EROPS', 'EROPS AC', 'EROPS w AC'] ] = enclosure_dummy[ ['OROPS','EROPS', 'EROPS AC', 'EROPS w AC'] ]
df.head()
```

```python
df[['SalePrice','age','BL','MG','SSL','TEX','WL','ProductGroupRank','EROPS', 'OROPS']].corr()
```

```python
y = df['SalePrice'][(df.age <= 30) & (df.age >= 0) & df.Enclosure.isin(['OROPS', 'EROPS', 'EROPS w AC'])]
```

```python
#X = df1[['SalesYear','YearMade','age','BL','MG','SSL','TEX','TTT','WL']][(df.age <= 30) & (df.age >= 0)]
X = df[['age','BL','MG','SSL','TEX','WL','ProductGroupRank','EROPS', 'OROPS']][(df.age <= 30) & (df.age >= 0) & df.Enclosure.isin(['OROPS', 'EROPS', 'EROPS w AC'])]
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
scalar = StandardScaler().fit(X_train)
X_train = scalar.transform(X_train)
X_test = scalar.transform(X_test)
```

## Basic Linear Reg

```python
model = LinearRegression().fit(X_train,y_train)
r2_score(y_test,model.predict(X_test))
```

```python
log_diff = np.log(model.predict(X_test)+1) - np.log(y_test+1)
np.sqrt(np.mean(log_diff**2))
```

## Grid Search

```python
params = {'alpha': np.logspace(-4, 5)}
gs = GridSearchCV(Ridge(), param_grid=params, cv=10, n_jobs=5).fit(X_train, y_train)
gs
```

```python
gs.best_estimator_
```

```python
gs.best_score_
```

```python
r2_score(y_test,gs.predict(X_test))
```

```python
log_diff = np.log(gs.predict(X_test)+1) - np.log(y_test+1)
np.sqrt(np.mean(log_diff**2))
```
